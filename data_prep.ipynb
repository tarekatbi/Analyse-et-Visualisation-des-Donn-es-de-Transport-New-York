```python
# Harmonisation des colonnes pour chaque dataset
yellow_data.rename(columns={
    'pickup_longitude': 'pickup_lon',
    'pickup_latitude': 'pickup_lat',
    'dropoff_longitude': 'dropoff_lon',
    'dropoff_latitude': 'dropoff_lat',
    'pickup_datetime': 'pickup_time',
    'dropoff_datetime': 'dropoff_time',
    'trip_distance': 'distance',
    'total_amount': 'total_fare'
}, inplace=True)

green_data.rename(columns={
    'Pickup_longitude': 'pickup_lon',
    'Pickup_latitude': 'pickup_lat',
    'Dropoff_longitude': 'dropoff_lon',
    'Dropoff_latitude': 'dropoff_lat',
    'pickup_datetime': 'pickup_time',
    'dropoff_datetime': 'dropoff_time',
    'Trip_distance': 'distance',
    'Total_amount': 'total_fare'
}, inplace=True)

vtc_data.rename(columns={
    'pickup_datetime': 'pickup_time',
    'locationID': 'pickup_location_id'
}, inplace=True)

# Vérification des colonnes renommées
print("Yellow Data Columns (after rename):", yellow_data.columns.tolist())
print("Green Data Columns (after rename):", green_data.columns.tolist())
print("VTC Data Columns (after rename):", vtc_data.columns.tolist())
```

```python
# Colonnes essentielles
essential_columns = ['pickup_time', 'pickup_lon', 'pickup_lat']

# Vérification
for name, dataset in [("Yellow", yellow_data), ("Green", green_data), ("VTC", vtc_data)]:
    print(f"\nColonnes manquantes dans {name}:")
    missing_columns = [col for col in essential_columns if col not in dataset.columns]
    if missing_columns:
        print(f" - {missing_columns}")
    else:
        print(" - Toutes les colonnes essentielles sont présentes.")
```

```python
# Conversion des colonnes temporelles
for dataset, name in [(yellow_data, "Yellow"), (green_data, "Green"), (vtc_data, "VTC")]:
    try:
        dataset['pickup_time'] = pd.to_datetime(dataset['pickup_time'], errors='coerce')
        if 'dropoff_time' in dataset.columns:
            dataset['dropoff_time'] = pd.to_datetime(dataset['dropoff_time'], errors='coerce')
    except Exception as e:
        print(f"Erreur dans la conversion des dates pour {name}: {e}")

# Vérification des plages temporelles
for dataset, name in [(yellow_data, "Yellow"), (green_data, "Green"), (vtc_data, "VTC")]:
    print(f"\n{name} Dataset Date Range:")
    print(f" - Pickup Min: {dataset['pickup_time'].min()}")
    print(f" - Pickup Max: {dataset['pickup_time'].max()}")
```

```python
import pandas as pd

# Définir les dates actuelles et cibles
current_min_date = pd.to_datetime("2015-04-01 21:37:57")  # Date actuelle minimale
target_min_date = pd.to_datetime("2015-01-01 21:37:57")  # Nouvelle date minimale
date_shift = target_min_date - current_min_date  # Calcul du décalage

# Appliquer le décalage aux colonnes des dates
green_data['pickup_time'] = pd.to_datetime(green_data['pickup_time']) + date_shift
green_data['dropoff_time'] = pd.to_datetime(green_data['dropoff_time']) + date_shift

# Vérifier les nouvelles plages de dates
print("Nouvelle plage des dates (pickup):")
print(f"Min: {green_data['pickup_time'].min()}")
print(f"Max: {green_data['pickup_time'].max()}")
```

```python
import pandas as pd

# S'assurer que pickup_datetime est au bon format
vtc_data['pickup_time'] = pd.to_datetime(vtc_data['pickup_time'])

# Définir la plage de dates souhaitée
start_date = pd.to_datetime("2015-01-01")
end_date = pd.to_datetime("2015-01-18")

# Filtrer les données entre ces dates
filtered_vtc_data = vtc_data[(vtc_data['pickup_time'] >= start_date) & (vtc_data['pickup_time'] <= end_date)]

# Vérifier les premières lignes du dataframe filtré
display(filtered_vtc_data.head())
display(len(filtered_vtc_data))

# Sauvegarder le dataset filtré (optionnel)
# filtered_vtc_data.to_parquet("filtered_vtc_data.parquet")
```

```python
import pandas as pd
from shapely import wkt

zones['coordinates'] = zones['the_geom'].apply(wkt.loads)  # Convertir en objets shapely
zones['pickup_longitude'] = zones['coordinates'].apply(lambda x: x.centroid.x)
zones['pickup_latitude'] = zones['coordinates'].apply(lambda x: x.centroid.y)

print(zones[['LocationID', 'pickup_longitude', 'pickup_latitude']].head())

# Joindre les données des coordonnées aux données VTC en utilisant 'pickup_location_id'
vtc_data_with_coords = filtered_vtc_data.merge(zones[['LocationID', 'pickup_longitude', 'pickup_latitude']], 
                                               how='left', 
                                               left_on='pickup_location_id', 
                                               right_on='LocationID')

# Sauvegarder le dataframe enrichi (optionnel)
vtc_data_with_coords.to_parquet("vtc_data_with_coordinates.parquet")
```

```python
# Fonction pour nettoyer et afficher un résumé du dataset vtc_data_with_coords
def clean_and_generate_summary_vtc(dataset):
    # Colonnes à supprimer
    columns_to_drop = ['month', 'date', 'time', 'day', 'Pickup_date']
    
    # Supprimer les colonnes spécifiées
    dataset_cleaned = dataset.drop(columns=columns_to_drop, axis=1)
    
    # Afficher un résumé des données nettoyées
    print("--- vtc_data_with_coords Summary (Cleaned) ---")
    print(f"Shape: {dataset_cleaned.shape}")
    print(f"Columns: {dataset_cleaned.columns}")
    print(f"Data Types: {dataset_cleaned.dtypes}")
    print(f"Missing Values: \n{dataset_cleaned.isnull().sum()}")
    print(f"Sample Data (First 5 rows): \n{dataset_cleaned.head()}")

    # Sauvegarder le résumé dans un fichier CSV
    summary_filename = "vtc_data_summary_cleaned.csv"
    dataset_cleaned.to_csv(summary_filename, index=False)
    print(f"Summary saved as {summary_filename}")

    # Sauvegarder le dataset nettoyé dans un fichier Parquet
    dataset_cleaned.to_parquet("vtc_data_cleaned.parquet")
    print("Cleaned vtc_data has been saved as vtc_data_cleaned.parquet")

    return dataset_cleaned

# Nettoyer et générer le résumé du dataset vtc_data_with_coords
vtc_data_cleaned = clean_and_generate_summary_vtc(vtc_data_with_coords)
```
